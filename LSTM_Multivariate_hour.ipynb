```python
import numpy as np
import pandas as pd

import tensorflow as tf
```

```python
data = pd.read_csv('daily_rv_hour_data.csv')
```

```python
data
```

```python
data = data.rename(columns={"Unnamed: 0": "Date"})
```

```python
import datetime

def str_to_datetime(s):
  split = s.split('/')
  month, day, year = int(split[0]), int(split[1]), int(split[2])
  return datetime.datetime(year=year, month=month, day=day)

datetime_object = str_to_datetime('1/10/2024')
datetime_object
```

```python
data['Date'] = data['Date'].apply(str_to_datetime)
data['Date']
```

```python
data.index = data.pop('Date')
data
```

```python
def df_to_X_Y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [r for r in df_as_np[i:i+window_size]]
    X.append(row)
    label = [df_as_np[i+window_size]]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
X, Y = df_to_X_Y(data)
X.shape, Y.shape
```

```python
q_80 = int(len(data.index) * .8)
q_90 = int(len(data.index) * .9)

X_train, Y_train =  X[:q_80], Y[:q_80]

X_val, Y_val =  X[q_80:q_90], Y[q_80:q_90]
X_test, Y_test =  X[q_90:], Y[q_90:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers

model = Sequential([layers.Input((5, 100)),
                    layers.LSTM(64),
                    layers.Dense(32, activation='relu'),
                    layers.Dense(100)])

model.summary()
```

```python
from keras import ops

def qlike(y_true, y_pred):
    sum = ops.log(y_pred+1e-16) + y_true/y_pred
    return ops.mean(sum, axis=-1)  # Note the `axis=-1`
```

```python
model.compile(loss='mse', 
              optimizer=Adam(learning_rate=0.001),
              metrics=['mean_absolute_error'])
```

```python
model.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=100)
```
